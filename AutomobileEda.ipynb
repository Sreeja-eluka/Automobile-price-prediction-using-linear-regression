```python
df_gptest = df[['drive-wheels','body-style','price']]
grouped_test1 = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean()
grouped_test1
```

```python
grouped_pivot = grouped_test1.pivot(index='drive-wheels',columns='body-style')
grouped_pivot
```

```python
grouped_pivot = grouped_pivot.fillna(0)
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

```python
# remove outliers
df_clean=df.copy(deep=True)

```

```python
df_clean.columns
```

```python
df_clean.drop(df_clean[df_clean['compression-ratio']>12].index,
              inplace=True)
df_clean.drop(df_clean[df_clean['peak-rpm']>6000].index,
              inplace=True)
df_clean.drop(df_clean[df_clean['highway-L/100km']>13].index,
              inplace=True)
```

```python
df_clean.shape
```

```python
fig,ax=plt.subplots(figsize=(15,10))
sns.heatmap(df.corr(),cmap='viridis_r',annot=True)
```

```python
from scipy import stats
correlation_table=pd.DataFrame()
correlation_table['features']=numerical_col
coeffs=[]
p_vs=[]
for col in numerical_col:
    coeff,p=stats.pearsonr(df_clean[col],df_clean['price'])
    coeffs.append(coeff)
    p_vs.append(p)
correlation_table['pearson_correlation']=coeffs
correlation_table['p_value']=p_vs
```

```python
correlation_table
```

```python
correlation_table[correlation_table['pearson_correlation']>0.4]#strong positive correlation
```
